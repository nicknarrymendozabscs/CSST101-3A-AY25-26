```python
# Get user preferences
budget = input("What is your budget? (low/medium/high): ").lower()
climate = input("Preferred climate? (hot/cold/moderate): ").lower()
activity = input("Preferred activity? (relaxing/adventure/culture/nature): ").lower()
duration = input("Travel duration? (short/medium/long): ").lower()

# Rule-based logic

if budget == "high" and climate == "moderate" and activity == "culture":
    suggestion = "Japan – A blend of modern and traditional culture"
elif budget == "medium" and climate == "cold" and activity == "adventure":
    suggestion = "South Korea – Winter sports and cultural sites"
elif budget == "high" and climate == "moderate" and activity == "relaxing":
    suggestion = "France – Wine regions and countryside relaxation"
elif budget == "medium" and climate == "hot" and activity == "shopping":
    suggestion = "Singapore – Urban exploration and shopping"
elif budget == "medium" and climate == "cold" and activity == "nature":
    suggestion = "Canada – National parks, lakes, and mountain adventures"
else:
    suggestion = "Try researching more destinations based on your preferences!"

# Output
print("\nRecommended Travel Destination:", suggestion)
```
